```python
import pandas
import numpy as np
from sklearn.neighbors import KNeighborsRegressor
from sklearn.cross_validation import KFold, cross_val_score
from sklearn import preprocessing
from sklearn.datasets import load_boston
# get boston data
df = load_boston()

# prepare data training input samples  
X = np.array(df['data'])
y = np.array(df['target'])

kf = KFold(len(X), n_folds=5, random_state=42)
# normalize features
X_scaled = preprocessing.scale(X)

# only for usefull
j = 0
for i in np.linspace(1.0, 10.0, num=200):
    j += 1
    neigh = KNeighborsRegressor(n_neighbors=5, weights='distance', p = i, metric='minkowski')
    print('Step: ' + str(j), end = ', ')
    print('p: ' + str(i) + ', ', end = '\t')
    print('Result: ', end = ' ')
    print(cross_val_score(neigh, X_scaled, y, scoring='mean_squared_error', cv=kf, n_jobs=-1).mean())
# We can see that max it's -26.975 and p = 1.4
```
