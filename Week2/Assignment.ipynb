```python
import pandas
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cross_validation import KFold, cross_val_score
from sklearn import preprocessing

#get data
df = pandas.read_csv('https://raw.githubusercontent.com/rasbt/pattern_classification/master/data/wine_data.csv', header=None)

# prepare data training input samples (from https://archive.ics.uci.edu/ml/datasets/Wine)
df.columns=['Class', 'Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash', 'Magnesium', 'Total phenols', 'Flavanoids'
              , 'Nonflavanoid phenols', 'Proanthocyanins', 'Color intensity', 'Hue', 'OD280/OD315 of diluted wines', 'Proline']
  
X = np.array(df[['Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash', 'Magnesium', 'Total phenols', 'Flavanoids'
              , 'Nonflavanoid phenols', 'Proanthocyanins', 'Color intensity', 'Hue', 'OD280/OD315 of diluted wines', 'Proline']])
y = np.array(df['Class'])

kf = KFold(len(X), n_folds=5, random_state=42)
# normalize features
X_scaled = preprocessing.scale(X)

max_count_of_neighbors = 50
for i in range(1, max_count_of_neighbors+1):
    neigh = KNeighborsClassifier(i)
    print('Count of neighbors: ' + str(i) + ':')
    
    print('Prediction average (without scale): ', end = ' ')
    print(round(cross_val_score(neigh, X, y,  cv=kf, n_jobs=-1).mean(), 3))
    
    print('Prediction average (with scale): ', end = ' ')
    print(round(cross_val_score(neigh, X_scaled, y,  cv=kf, n_jobs=-1).mean(), 3))
    print('_________________________________')
    print()
    
# We can see that max result without scale it's 0.63 and k=1. With scale 0.93 and k = 16
```
